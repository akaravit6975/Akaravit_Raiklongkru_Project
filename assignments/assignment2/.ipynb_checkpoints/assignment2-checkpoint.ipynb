### Task 2: Retrieving data from MarketStack (1p)

Use the [MarketStack API](https://docs.apilayer.com/marketstack/docs/api-documentation?utm_source=dashboard&utm_medium=Referral) to investigate how the FAANG stocks performed in March 2026. The FAANG companies and their ticker symbols are: Facebook/Meta (META), Apple (AAPL), Amazon (AMZN), Netflix (NFLX), and Google/Alphabet (GOOG).

a) Retrieve End-of-Day (eod) data for these five stocks for March 2026. Note: Each ticker in a request counts as a separate API call against your monthly limit, so plan your requests accordingly. Retrieving the data once should take only 5% of the monthly free API calls.

b) Convert the retrieved JSON data into a pandas DataFrame. Print the first 5 rows.

c) Compute summary statistics for the closing prices grouped by stock. Also calculate the return percentage for each stock over the month (i.e. the percentage change from the first to the last trading day). Print out the computed numbers or a DataFrame that contains the numbers, and report which stock performed best and which worst in March 2026.

Hint: The API documentation helps on how to use it, but also using AI is recommended to help with formulating the GET request if you get stuck. Microsoft Copilot is usually very bad so I suggest using Claude, ChatGPT or Le Chat.

#### 2a) Retrieve EOD data for the five stocks for Mar 2026

````python
import requests
import pandas as pd

url = "http://api.marketstack.com/v1/eod?symbols=META,AAPL,AMZN,NFLX,GOOG"

params = {
    'access_key': api_key, 
    'date_from': '2026-03-01',
    'date_to': '2026-03-31',
    'limit': 1000
}

response = requests.get(url, params=params)
data = response.json()
````

#### 2b) Convert to DataFrame and print first 5 rows

````python
df = pd.DataFrame(data['data'])

# Tidy up the date column and sort
df['date'] = pd.to_datetime(df['date']).dt.date
df = df.sort_values(['symbol', 'date']).reset_index(drop=True)

print("First 5 rows:")
df.head()
````

#### 2c) Summary Statistics

````python
# 1. Compute summary statistics for closing prices grouped by stock
summary_stats = df.groupby('symbol')['close'].describe()
print("--- Summary Statistics for Closing Prices ---")
display(summary_stats)

# 2. Calculate the return percentage for each stock over the month
first_close = df.groupby('symbol')['close'].first()
last_close = df.groupby('symbol')['close'].last()

# Calculate percentage change
return_pct = ((last_close - first_close) / first_close) * 100

# Put the results into a clean DataFrame for easy viewing
returns_df = pd.DataFrame({
    'First Close': first_close,
    'Last Close': last_close,
    'Return (%)': return_pct
}).sort_values('Return (%)', ascending=False)

print("\n--- March 2026 Return Percentages ---")
display(returns_df)

# 3. Report the best and worst performers
best_stock = returns_df['Return (%)'].idxmax()
best_return = returns_df['Return (%)'].max()

worst_stock = returns_df['Return (%)'].idxmin()
worst_return = returns_df['Return (%)'].min()

print(f"\nBest Performer in March 2026: {best_stock} ({best_return:.2f}%)")
print(f"Worst Performer in March 2026: {worst_stock} ({worst_return:.2f}%)")
````

#### 6c) Revenue by Product Category (SQL + Plot)

````python
query_6c = """
    SELECT 
        cat.CategoryName,
        SUM(od.UnitPrice * od.Quantity) AS TotalRevenue
    FROM "Order Details" od
    JOIN Products p ON od.ProductID = p.ProductID
    JOIN Categories cat ON p.CategoryID = cat.CategoryID
    GROUP BY cat.CategoryName
    ORDER BY TotalRevenue DESC;
"""

# Execute the query and load into a DataFrame
df_6c = pd.read_sql_query(query_6c, conn)

print("--- Total Revenue by Category ---")
display(df_6c.head())

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=df_6c, 
    x='TotalRevenue', 
    y='CategoryName', 
    hue='CategoryName',
    palette='rocket',
    legend=False
)

plt.title('Total Revenue by Product Category', fontsize=14, fontweight='bold')
plt.xlabel('Total Revenue ($)', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.tight_layout()
plt.show()
````

#### 7c) Execution Timing

````python
import time

start_sql = time.time()

query_sql_fast = """
    SELECT 
        cat.CategoryName,
        SUM(od.UnitPrice * od.Quantity) AS TotalRevenue
    FROM "Order Details" od
    JOIN Products p ON od.ProductID = p.ProductID
    JOIN Categories cat ON p.CategoryID = cat.CategoryID
    GROUP BY cat.CategoryName;
"""
df_sql_result = pd.read_sql_query(query_sql_fast, conn)

end_sql = time.time()
sql_duration = end_sql - start_sql

# 2. Timing the Pandas Approach
start_pandas = time.time()

# Step A: Retrieve all raw tables
df_od = pd.read_sql_query('SELECT * FROM "Order Details";', conn)
df_p = pd.read_sql_query('SELECT * FROM Products;', conn)
df_c = pd.read_sql_query('SELECT * FROM Categories;', conn)

# Step B: Perform the merges in Python
# We add suffixes because both tables have a 'UnitPrice' column
merged_1 = pd.merge(df_od, df_p, on='ProductID', how='inner', suffixes=('_od', '_p'))
merged_2 = pd.merge(merged_1, df_c, on='CategoryID', how='inner')

# Step C: Replicate the math and grouping
merged_2['LineTotal'] = merged_2['UnitPrice_od'] * merged_2['Quantity']
df_pandas_result = merged_2.groupby('CategoryName')['LineTotal'].sum().reset_index()

end_pandas = time.time()
pandas_duration = end_pandas - start_pandas

# 3. Print Results
print(f"SQL Execution Time:    {sql_duration:.5f} seconds")
print(f"Pandas Execution Time: {pandas_duration:.5f} seconds") ### SQL is faster in terms of execution
````

### Task 8: Wide and long format (1p)

You are given two datasets. The first is in wide format:

```python
wide_data = pd.DataFrame({
    "store": ["Helsinki", "Tampere", "Turku"],
    "Q1_sales": [12000, 8500, 9200],
    "Q2_sales": [13500, 9100, 9800],
```

The second is in long format:

```python
long_data = pd.DataFrame({
    "month": ["Jan", "Jan", "Feb", "Feb", "Mar", "Mar"],
    "product": ["Laptop", "Phone", "Laptop", "Phone", "Laptop", "Phone"],
    "units_sold": [45, 120, 50, 110, 55, 130]
})
```

a) Convert `wide_data` from wide to long format using `pd.melt()`. Print the DataFrame before and after conversion.

b) Convert `long_data` from long to wide format using `.pivot()`. Print the DataFrame before and after conversion.

c) Convert both transformed datasets back to their original format. Print the datasets before and after to confirm they match the originals.

#### 8a) Wide to Long using pd.melt()

````python
# Initialize the given datasets
wide_data = pd.DataFrame({
    "store": ["Helsinki", "Tampere", "Turku"],
    "Q1_sales": [12000, 8500, 9200],
    "Q2_sales": [13500, 9100, 9800],
    "Q3_sales": [11800, 8800, 10100],
    "Q4_sales": [15000, 10200, 11500]
})

long_data = pd.DataFrame({
    "month": ["Jan", "Jan", "Feb", "Feb", "Mar", "Mar"],
    "product": ["Laptop", "Phone", "Laptop", "Phone", "Laptop", "Phone"],
    "units_sold": [45, 120, 50, 110, 55, 130]
})

# --- 8a) Convert wide_data to long format using pd.melt() ---
print("8A: WIDE TO LONG")
print("Before (Original Wide):")
display(wide_data)

wide_to_long = pd.melt(
    wide_data, 
    id_vars=["store"],          # The column to keep as an identifier
    var_name="quarter",         # Name for the new column containing the old column headers
    value_name="sales"          # Name for the new column containing the actual numbers
)

print("\nAfter (Transformed to Long):")
display(wide_to_long)
````

#### 8b) Convert long_data to wide format using .pivot()

````python
print("8B: LONG TO WIDE")
print("Before (Original Long):")
display(long_data)

long_to_wide = long_data.pivot(
    index="month",              # The column to use as row labels
    columns="product",          # The column whose unique values become new column headers
    values="units_sold"         # The column containing the data to fill the grid
).reset_index()                 # Flattens the index back to a standard DataFrame

# Optional: Clear the generic axis name 'product' that pivot leaves behind for a cleaner look
long_to_wide.columns.name = None 

print("\nAfter (Transformed to Wide):")
display(long_to_wide)
````

#### 8c) Revert to Originals

````python
print("8C: REVERTING DATASETS")

# 1. Revert 'wide_to_long' back to its original wide format using .pivot()
reverted_wide = wide_to_long.pivot(
    index="store", 
    columns="quarter", 
    values="sales"
).reset_index()

reverted_wide.columns.name = None

# Reorder the columns to match the exact layout of the original 'wide_data'
reverted_wide = reverted_wide[['store', 'Q1_sales', 'Q2_sales', 'Q3_sales', 'Q4_sales']]

print("Reverted Wide Data (Matches Original):")
display(reverted_wide)


# 2. Revert 'long_to_wide' back to its original long format using pd.melt()
reverted_long = pd.melt(
    long_to_wide,
    id_vars=["month"],
    var_name="product",
    value_name="units_sold"
)

month_mapping = {'Jan': 1, 'Feb': 2, 'Mar': 3}
reverted_long['month_num'] = reverted_long['month'].map(month_mapping)

# Sort by the month number, then product, then drop the temporary sorting column
reverted_long = reverted_long.sort_values(['month_num', 'product']).drop('month_num', axis=1).reset_index(drop=True)

print("\nReverted Long Data (Matches Original):")
display(reverted_long)
````
